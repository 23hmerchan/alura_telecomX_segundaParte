```python
import pandas as pd
import json

# Cargar el archivo JSON
with open('TelecomX_Data.json') as f:
    data_telecom = json.load(f)

# Aplanar el JSON en un solo DataFrame plano
df_flat = pd.json_normalize(data_telecom, sep='_')

# Mostrar las primeras filas
df_flat
```

#🔧 Transformación

```python
data_telecomx_final['customerID']=data_telecomx_final['customerID'].astype(str)

```

```python
data_telecomx_final['account_Charges_Total']=pd.to_numeric(data_telecomx_final['account_Charges_Total'],errors='coerce')
```

```python
data_telecomx_final.dtypes  # Verifica tipos exactos

# O verifica tipo de un valor individual
type(data_telecomx_final['phone_MultipleLines'][20])  # Te mostrará <class 'str'>
```

```python
cols_categorical = [
    'customer_gender', 'phone_MultipleLines',
    'internet_InternetService', 'account_Contract',
    'account_PaymentMethod'
]

for col in cols_categorical:
    data_telecomx_final[col] = data_telecomx_final[col].astype('category')
```

```python
cols_with_nan = [
    'internet_OnlineSecurity', 'internet_OnlineBackup',
    'internet_DeviceProtection', 'internet_TechSupport',
    'internet_StreamingTV', 'internet_StreamingMovies'
]
data_telecomx_final[cols_with_nan] = data_telecomx_final[cols_with_nan].fillna(False)
```

```python
data_telecomx_final['has_internet'] = data_telecomx_final['internet_InternetService'] != 'No'
```

```python
data_telecomx_final.dropna(inplace=True)
```

#📊 Carga y análisis

```python
datos_telecomx=pd.read_csv('/content/telecomx_transformado_total.csv')
datos_telecomx.head()
```

```python
datos_telecomx_churn=datos_telecomx.value_counts('Churn')
datos_telecomx_churn.head()
```

```python
#porcentaje
datos_telecomx_churn_porc=datos_telecomx['Churn'].value_counts(normalize=True)*100
datos_telecomx_churn_porc
```

```python
import matplotlib.pyplot as plt

# Etiquetas y colores
labels = ['Se quedaron', 'Se fueron']
colors = ['#66b3ff', '#ff6666']

# Gráfico de pastel con porcentajes
plt.figure(figsize=(6, 6))
plt.pie(datos_telecomx_churn_porc, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90, explode=(0, 0.1))
plt.title('Porcentaje de Clientes que se fueron vs se quedaron')
plt.axis('equal')
plt.show()
```

```python
churn_por_contrato = datos_telecomx.groupby('account_Contract')['Churn'].value_counts(normalize=True).unstack() * 100
churn_por_contrato
```

```python
churn_por_contrato.plot(kind='bar', stacked=True, color=['blue','red'], figsize=(6, 4))
plt.title('Porcentaje de Churn según Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Porcentaje de Clientes')
plt.legend(title='Churn', labels=['Se quedaron', 'Se fueron'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Según la grafica los la cantidad de las personas que se fueron estan relacionados por el contrato de mes a mes, esto podemos describirlo como una inestabilidad en el contrato

```python
servicios = [
    'internet_TechSupport',
    'internet_OnlineSecurity',
    'internet_OnlineBackup',
    'internet_DeviceProtection',
    'internet_StreamingTV',
    'internet_StreamingMovies'
]

for servicio in servicios:
    print(f'\n📌 {servicio}')
    display(pd.crosstab(datos_telecomx[servicio], datos_telecomx['Churn'], normalize='index') * 100)
```

Conclusion general del analisis
Se analizaron 7,032 clientes, de los cuales 1,869 se dieron de baja, lo que representa una tasa de churn del 26.6%, mientras que 5,163 permanecen activos, equivalente al 73.4%. Esto indica que más de 1 de cada 4 clientes abandona el servicio, lo cual representa una alerta importante en términos de retención. La mayor concentración de clientes que se fueron corresponde al segmento con contrato “Month-to-month”, donde el 42.7% abandonó el servicio, en contraste con tasas significativamente más bajas en los contratos de mayor permanencia: 11.5% en contratos de un año y apenas 3.4% en contratos de dos años. Estos datos evidencian que la modalidad de pago mensual es altamente inestable y se consolida como un foco crítico de evasión.
